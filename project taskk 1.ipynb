```python
import pandas as pd

df = pd.read_excel("Dataset for Data Analytics.xlsx")

print("Missing values per column:")
print(df.isnull().sum())

df['CouponCode'] = df['CouponCode'].fillna('None')

print("\nDuplicate OrderIDs:", df['OrderID'].duplicated().sum())
print("Duplicate rows:", df.duplicated().sum())

mismatch = (df['Quantity'] * df['UnitPrice']).round(2) != df['TotalPrice']
print("TotalPrice mismatches:", mismatch.sum())

date_check = pd.api.types.is_datetime64_any_dtype(df['Date'])
print("Dates valid datetime format:", date_check)

df.to_excel("Cleaned_Dataset.xlsx", index=False)
print("\nCleaned file saved successfully!")
```

```python
# Fix TotalPrice mismatches
mismatch = (df['Quantity'] * df['UnitPrice']).round(2) != df['TotalPrice']
print("Mismatched rows:", mismatch.sum())
print(df[mismatch][['OrderID','Quantity','UnitPrice','TotalPrice']])

# Correct the TotalPrice
df['TotalPrice'] = (df['Quantity'] * df['UnitPrice']).round(2)

# Save final cleaned file
df.to_excel("Cleaned_Dataset.xlsx", index=False)
print("\nFinal corrected file saved!")
```
